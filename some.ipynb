```python
import matplotlib.pyplot as plt
import numpy as np

# Данные
U_in = np.array([100, 200, 500, 1000, 1100, 1200, 1300, 1500, 2000, 3000])
U_out = np.array([152, 304, 762, 1450, 1500, 1600, 1675, 1840, 2230, 2995])

# Создание графика
plt.figure(figsize=(10, 6))

# Только синяя сплошная линия
plt.plot(U_in, U_out, color='blue', marker='o', linestyle='-', linewidth=2, label='Uвых(Uвх)')

# Настройка осей
plt.xlabel('Uвх, мВ', fontsize=12)
plt.ylabel('Uвых, мВ', fontsize=12)
plt.title('Амплитудная характеристика каскада с ОЭ (Вариант 17)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper left')

# Пределы осей
plt.xlim(0, 3000)
plt.ylim(0, 3500)

plt.tight_layout()
plt.show()
```
